```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
X = Coaches[['StadSize', 'MedianConfSal', 'Ratio', 'TrueRank']]
y = Coaches['SchoolPay']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
m = mean_absolute_error(y_test, pred)
print("Mean Absolute Error:", m)
```

```python
data = [[60000, 3000000.0, 0.7, 4]]
predicted_salary = model.predict(data)
print("Predicted Coach Salary: $", predicted_salary[0])
```
